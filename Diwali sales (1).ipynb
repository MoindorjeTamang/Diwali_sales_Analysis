# Exploratory Data analysis

```python
# Creating count plot
ax = sns.countplot(x='Gender', data=df)

for bar in ax.containers: # Using for loop for implement the data labels values
    ax.bar_label(bar)
```

```python
# Using groupby filter for the filtering the data as per columns needs
Barchart = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=Barchart)
pyp.show()

#for i in Barchart.containers:
 #   Barchart.bar_label(i)
```

* Most of the product are buying by Female as compared to Male

```python
Agechart = sns.countplot(x='Age Group', hue='Gender',data=df)

for i in Agechart.containers:
    Agechart.bar_label(i)
```

* Most of the buyer are Female at the age of between 26 to 35.
* Male have also the same age group catogories, who are buying highest product.

```python
Agegroup = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount',data=Agegroup)
pyp.show()
```

* In the above chart most of the spending age group is 26-35 yrs customers

```python
df.columns
```

```python
# Top 10 orders state names

State1 = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=State1)
pyp.show()
```

* Most of the order state is Uttar pradesh
* Second highest order state is Maharashtra

```python
# Top 10 orders state names

Amount1 = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=Amount1)
pyp.show()
```

* In terms of the total value of amount, Uttar Pradesh holds the top position, with Maharashtra closely following in second place.

* In the above chart if you compared with order to Amount figure, You have cleared seen the kerela is not existing under top 10 list on the basis of amount.

```python
# Drilling the Marital Status wise
MS = sns.countplot(x='Marital_Status', hue='Gender', data=df)

# Annotate each bar with its count value
for container in MS.containers:
    MS.bar_label(container)

# Set x-axis label
pyp.xlabel('Marital Status')

# Show the plot
pyp.show()
```

* Purchasing activity is higher among married women in comparison to single women

```python
MTS = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc={'figure.figsize':(10,5)})
data2 = sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=MTS,)



pyp.xlabel('Marital Status')
pyp.ylabel('Total Amount')
pyp.legend(title='Gender')
pyp.show()
```

* Married women spending more money for buying our product as compared to un-married
